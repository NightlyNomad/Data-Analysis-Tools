# 1. Демонстрация 12 любых функций / методов / операторов из документации по numpy в блокноте.

```python
print("1. Создание массива: \n")
array = np.array([[66], [8]])
print(array)

print("2. Трансформация каждого элемента с использованием log(1+x) функции: \n")
log_transformed_array = np.log1p(array)
print(log_transformed_array)

print("3. Генерация матрицы: \n")
matrix = np.matrix('11 -2 ; 32 6 ')
print(matrix)

print("4. Решение линейных матричных уравнений или системы линейных скалярных уравнений: \n")
solution = np.linalg.solve(matrix, array)
print(solution)

print("5. Возведение в третью степень: \n")
power_x = np.power(solution, 3)
print(power_x)

print("6. Вычисление 0.5 квантиля: \n")
quantile_solution = np.quantile(solution, 0.5)
print(quantile_solution)

print("7. Конкатенация двух массивов: \n")
conc_array = np.concatenate((array, log_transformed_array), axis = 1)
print(conc_array)

print("8. Получение размера массива: \n")
print(conc_array.size)

print("9. Получение размерности массива: \n")
print(conc_array.ndim)

print("10. Сумма элементов массива: \n")
print(conc_array.sum(axis=0))

print("11. Получение уникальных значений в массиве: \n")
unique_values = np.unique(np.array([1, 2, 67, 4, 2, 54, 3, 1]))
print(unique_values)

print("12. Дисперсия массива: \n")
variance = np.var(np.array([1, 2, 67, 4, 2, 54, 3, 1]))
print(variance)
```

# 2. Генерация OLAP-куба, с фиксацией продажи товаров в розничной сети по магазинам. 

Оси: 
- 0: ID группы товаров 
- 1: ID магазина 
- 2: день месяца, в котором осуществляются продажи (предположим у нас в OLAP-кубе собраны продажи за месяц). 

- Содержимое ячейки: продажи в тыс. рублей данной категории товаров в данном магазине в данный день.

```python
products = np.arange(0, 30, 1)
shops = np.arange(0, 10, 1)
days = np.arange(1, 31, 1)
cube = np.zeros([products.size, shops.size, days.size])

for p in range(products.size):
  for s in range(shops.size):
    for d in range(days.size):
      cube[p, s, d] = np.random.randint(10, 80)

print(cube)
```

# 3. Для получившегося OLAP-куба посчитать суммарные, средние, минимальные и максимальные продажи, зафиксировав значения по одной, двум или трём осям

```python
print("Общая сумма продаж всех товаров во всех магазинах за весь месяц: \n")
print(np.sum(cube[:,:,:]))

print("Сумма по продажам 1-5 чисел месяца для товара 9 по всем магазинам: \n")
print(np.sum(cube[9,:,[days[0:5]]]))

print("Минимум по продажам 10-20 чисел месяца для товара 16 в магазине 3: \n")
print(np.min(cube[16,3,[days[9:20]]]))

print("Минимум по продажам 15-20 чисел месяца для товара 9 в магазине 2: \n")
print(np.min(cube[9,2,[days[14:20]]]))

print("Стандартное отклонение по продажам 1-10 чисел месяца для товара 1 по всем магазинам: \n")
print(np.std(cube[1,:,[days[0:10]]]))

print("Максимум по продажам за все время для товара 12 в магазине 2: \n")
print(np.max(cube[12,2,:]))

print("Cреднее по продажам за все время для товара 1 в магазине 1: \n")
print(np.mean(cube[1,1,:]))
```

#### Мы можем наблюдать графическое представление о продаже каждого продукта в месяце

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(range(30), cube[0,1,:])
plt.plot(range(30), cube[1,1,:])
plt.plot(range(30), cube[2,1,:])
plt.plot(range(30), cube[3,1,:])
# etc, for simplify should write 'for' or 'while' cycle
plt.show()
```
